```python
import pandas as pd
url = 'https://raw.githubusercontent.com/klaytoncastro/idp-machinelearning/main/decision-tree/winequality-merged.csv'
arquivo = pd.read_csv(url)
arquivo.head()
```

```python
# Substituindo a varivel categrica por numrica
arquivo['color'] = arquivo['color'].replace('red', 0)
arquivo['color'] = arquivo['color'].replace('white', 1)
arquivo['color'] = arquivo['color'].astype('int')
arquivo.head()
```

```python
arquivo.info()
```

```python
# Definindo o limiar para corte de vinhos bons como os de qualidade >=7
import numpy as np
arquivo['worst'] = np.where(arquivo['quality'] < 7, 1, 0)
arquivo.head()
```

```python
#arquivo = arquivo.drop(['alcohol', free sulfur dioxide', total sulfur dioxide'], axis=1) # Teste removendo vrias colunas com baixa correlao

arquivo = arquivo.drop(['quality'], axis=1) #Removendo a coluna 'quality' para no interferir nos clculos do modelo para predio da varivel dependente 'worst'

arquivo.head()
```

```python
# Exibir o percentual de cada classe
print(arquivo['worst'].value_counts(normalize=True) * 100)
```

```python
# Definindo os atributos da funo de aprendizagem
y = arquivo['worst']
X = arquivo.drop('worst',axis = 1)
```

```python
from sklearn.model_selection import train_test_split

# Definindo os conjuntos de treino e teste, onde x  o conjunto de atributos (features que so nossas variveis preditoras) e y  a varivel alvo.
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=47)
```

```python
%%time
from skopt import BayesSearchCV
import xgboost as xgb

# Definindo o modelo
modelo = xgb.XGBClassifier(objective='binary:logistic')

# Parmetros para o Randomized Search
param_space = {
    'max_depth': [20, 50, 100],
    'n_estimators': [40, 400, 1200],
    'learning_rate': [0.11, 0.13, 0.19],
     'subsample': [1.0, 0.5, 0.2],
    'colsample_bytree': [0.2, 0.9, 1.0]
}


# Configurando o GridSearchCV
# grid_search = GridSearchCV(estimator=modelo, param_grid=param_grid, scoring='accuracy', cv=2, verbose=1)

bayes_search = BayesSearchCV(estimator=modelo, search_spaces = param_space, n_iter =10 , cv =2, verbose=1)

# Treinando o GridSearchCV
bayes_search.fit(x_train, y_train)

# Melhores parmetros e score
print("Melhores Parmetros:", bayes_search.best_params_)
print("Melhor score de acurcia:", bayes_search.best_score_)
```

```python
import xgboost as xgb
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Criando um pipeline para escalar os dados e aplicar o XGB
modelo = Pipeline([
    ('scaler', StandardScaler()),  # Normaliza os dados
    ('xgb', xgb.XGBClassifier(objective='binary:logistic', learning_rate=0.11, n_estimators=400, max_depth=50, subsample=0.5, colsample_bytree=0.9))  # Inicia o XGB
])

modelo = xgb.XGBClassifier(objective='binary:logistic', learning_rate=0.11, n_estimators=40, max_depth=50, subsample=1, colsample_bytree=0.7)
#modelo = xgb.XGBClassifier(objective='binary:logistic', learning_rate=0.096, n_estimators=53, max_depth=30, subsample=0.5, colsample_bytree=0.5)

# Treinando o modelo
modelo.fit(x_train, y_train)
```

```python
#Realizando a Predio
y_pred = modelo.predict(x_test)

#Verificao de Acurcia
resultado = modelo.score(x_test, y_test)
print ("Acurcia:", resultado)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

"""
Calculando e exibindo a matriz de confuso. A orientao padro  a seguinte:
[0,0]: Verdadeiros Negativos (VN) - Previses corretamente identificadas como negativas.
[0,1]: Falsos Positivos (FP) - Previses incorretamente identificadas como positivas.
[1,0]: Falsos Negativos (FN) - Previses incorretamente identificadas como negativas.
[1,1]: Verdadeiros Positivos (VP) - Previses corretamente identificadas como positivas.
"""

conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confuso:")
print(conf_matrix)
```

```python
"""
Calculando e exibindo as mtricas de classificao.
Se algumas classes tm muito mais amostras do que outras, isso pode influenciar o desempenho e confiabilidade do modelo.

O "support" refere-se  quantidade de ocorrncias da classe especfica no conjunto de dados, sendo til para verificar desbalanceamentos.
A "macro avg" calcula a mdia aritmtica das mtricas (preciso, recall, F1-score) para cada classe, sem considerar o nmero de instncias em cada classe (support).
A "weighted avg" calcula a mdia ponderada das mtricas para cada classe, considerando o nmero de instncias em cada classe (support).
"""
print("Relatrio de Classsificao:")
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Prever as probabilidades para os dados de teste.
# Note que estamos interessados nas probabilidades da classe positiva (1), ento usamos [:, 1].
y_probs = modelo.predict_proba(x_test)[:, 1]

# Calcular FPR, TPR, e limiares
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Calcular a AUC
roc_auc = auc(fpr, tpr)

# Plotar a curva ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
