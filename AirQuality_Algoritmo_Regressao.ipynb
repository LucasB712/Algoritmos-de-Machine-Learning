```python
from sklearn.linear_model import LinearRegression, RANSACRegressor
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=47)


models = {
    LinearRegression(),
    ExtraTreesRegressor(),
    RANSACRegressor(),
    RandomForestRegressor(),
    GradientBoostingRegressor(),
    KNeighborsRegressor(),
    SVR(kernel="rbf")
}

metrics = {
    "MAE": [],
    "MSE": [],
    "RMSE": [],
    "R2": []
}

for model in models:
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)

  mae = mean_absolute_error(y_test, y_pred)
  mse = mean_squared_error(y_test, y_pred)
  rmse = np.sqrt(mse)
  r2 = r2_score(y_test, y_pred)

  metrics["MAE"].append(mae)
  metrics["MSE"].append(mse)
  metrics["RMSE"].append(rmse)
  metrics["R2"].append(r2)

  print(model)
  print("MAE:", mae)
  print("MSE:", mse)
  print("RMSE:", rmse)
  print("R2:", r2)

  print("\n")

```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import randint
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np


# Definindo o regressor
reg = RandomForestRegressor()

# Definindo os hiperparâmetros para busca aleatória
param_dist = {
    "n_estimators": randint(10, 100),
    "max_depth": randint(1, 20),
}

# Criando a função para calcular métricas de regressão
def evaluate_regression(model, X, y_true):
    y_pred = model.predict(X)
    r2 = r2_score(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    return {'R^2': r2, 'MSE': mse, 'MAE': mae, 'RMSE' : rmse}

# Criando o objeto RandomizedSearchCV com R^2 como métrica
random_search = RandomizedSearchCV(reg, param_distributions=param_dist, n_iter=100, cv=2, scoring='r2', random_state=42)

# Realizando a busca aleatória
random_search.fit(X, y)

# Imprimindo os melhores hiperparâmetros encontrados
print("Melhores hiperparâmetros encontrados:")
print(random_search.best_params_)

# Calculando as métricas do melhor modelo
metrics = evaluate_regression(random_search.best_estimator_, X, y)

# Imprimindo as métricas
print("Métricas do melhor modelo:")
for metric, value in metrics.items():
    print(f"{metric}: {value}")
```
